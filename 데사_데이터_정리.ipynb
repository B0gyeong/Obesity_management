```python
selected_columns_df_dx_08 = df_dx_08[['ID','DW_WBT_pFT']]
selected_columns_df_dx_09 = df_dx_09[['ID','DW_WBT_pFT']]
selected_columns_df_dx_10 = df_dx_10[['ID','DW_WBT_pFT']]
selected_columns_df_dx_11 = df_dx_11[['ID','DW_WBT_pFT']]
```

```python
print(selected_columns_df_dx_08.shape)
print(selected_columns_df_dx_09.shape)
print(selected_columns_df_dx_10.shape)
print(selected_columns_df_dx_11.shape)
```

```python
selected_columns_df_08.rename(columns={'id': 'ID'}, inplace=True)
selected_columns_df_09.rename(columns={'id': 'ID'}, inplace=True)
selected_columns_df_10.rename(columns={'id': 'ID'}, inplace=True)
selected_columns_df_11.rename(columns={'id': 'ID'}, inplace=True)
```

```python
merged_df_08 = selected_columns_df_08.merge(selected_columns_df_dx_08, on='ID', how='inner')
merged_df_09 = selected_columns_df_09.merge(selected_columns_df_dx_09, on='ID', how='inner')
merged_df_10 = selected_columns_df_10.merge(selected_columns_df_dx_10, on='ID', how='inner')
merged_df_11 = selected_columns_df_11.merge(selected_columns_df_dx_11, on='ID', how='inner')
```

```python
print(merged_df_08.shape)
print(merged_df_09.shape)
print(merged_df_10.shape)
print(merged_df_11.shape)
```

```python
cleaned_df_08 = merged_df_08.dropna()
cleaned_df_09 = merged_df_09.dropna()
cleaned_df_10 = merged_df_10.dropna()
cleaned_df_11 = merged_df_11.dropna()
```

```python
cleaned_df_08 = cleaned_df_08.drop(columns=['ID'])
cleaned_df_09 = cleaned_df_09.drop(columns=['ID'])
cleaned_df_10 = cleaned_df_10.drop(columns=['ID'])
cleaned_df_11 = cleaned_df_11.drop(columns=['ID'])
```

```python
print(cleaned_df_08.shape)
print(cleaned_df_09.shape)
print(cleaned_df_10.shape)
print(cleaned_df_11.shape)
```

```python
import pandas as pd

frames = [cleaned_df_08, cleaned_df_09, cleaned_df_10, cleaned_df_11]
combined_df_total = pd.concat(frames, ignore_index=True)
```

```python
combined_df_total.shape
```

```python
print(combined_df_total.columns)
combined_df_total.head()
```

```python
#filtered_df = combined_df_total[combined_df_total['age'] >= 50]
filtered_df = combined_df_total

# 결과 출력
print(filtered_df)
```

```python
import math

filtered_df['wwi'] = filtered_df['HE_wc'] / filtered_df['HE_wt'].apply(math.sqrt)

# 결과 출력
print(filtered_df)
```

```python
filtered_df['ID'] = range(1, len(filtered_df) + 1)
filtered_df
```
